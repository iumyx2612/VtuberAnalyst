```python
# Get details information of videos

for i, videos in enumerate(video_ids):
    try:
        request_vd_detail = youtube.videos().list(
            part="snippet, statistics, contentDetails, liveStreamingDetails, topicDetails",
            id=videos,
            maxResults=50
        )
        video_detail = request_vd_detail.execute()

        # Extract data from the API response
        snippet = video_detail['items'][0]['snippet']
        statistics = video_detail['items'][0]['statistics']
        content_details = video_detail['items'][0]['contentDetails']
        live_streaming_details = video_detail['items'][0].get('liveStreamingDetails', {})
        topic_details = video_detail['items'][0].get('topicDetails', {})

        # Extract fields
        Video_ID = video_detail['items'][0]['id']
        Channel_ID = snippet['channelId']
        Video_Title = snippet['title']
        Published_At = snippet['publishedAt']
        Thumbnails = snippet['thumbnails']['default']['url']
        Comments = statistics.get('commentCount', 0)
        Likes = statistics.get('likeCount', 0)
        Views = statistics.get('viewCount', 0)
        Duration = content_details['duration']
        Start_Time = live_streaming_details.get('actualStartTime', '')
        End_Time = live_streaming_details.get('actualEndTime', '')
        Scheduled_Time = live_streaming_details.get('scheduledStartTime', '')
        Topic_Category = topic_details.get('topicCategories', '')

        # Create row and add to DataFrame
        row = [Video_ID, Channel_ID, Video_Title, Published_At, Thumbnails, Comments, Likes, Views, Duration,
               Start_Time, End_Time, Scheduled_Time, Topic_Category]
        df_video.loc[i] = row

    except Exception as e:
        print(f"Error processing video {videos}: {e}")
```
